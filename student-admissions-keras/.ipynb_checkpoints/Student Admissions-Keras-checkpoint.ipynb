```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


data=pd.read_csv('student_data.csv')

data.shape
```

```python
data
```

```python
def plot_points(data):
    X=np.array(data[['gre','gpa']])
    y=np.array(data['admit'])
    admitted=X[np.argwhere(y==1)]
    rejected=X[np.argwhere(y==0)]
    plt.scatter([s[0][0]for s in rejected],[s[0][1]for s in rejected],s=25,color='red',edgecolor='k')
    plt.scatter([s[0][0]for s in admitted],[s[0][1]for s in admitted],s=25,color='cyan',edgecolor='k')
    plt.xlabel('Test(GRE)')
    plt.ylabel('Test(GPA)')

plot_points(data)
plt.show()
```

```python
plt.scatter(data['gre'],data['gpa'])
plt.show()
```

```python
data_rank1=data[data['rank']==1]
data_rank2=data[data['rank']==2]
data_rank3=data[data['rank']==3]
data_rank4=data[data['rank']==4]

plot_points(data_rank1)
plt.title('rank1')
plt.show()

plot_points(data_rank2)
plt.title('rank2')
plt.show()

plot_points(data_rank3)
plt.title('rank3')
plt.show()

plot_points(data_rank4)
plt.title('rank4')
plt.show()


```

```python
data.head()
```

```python
one_hot_data=pd.concat([data,pd.get_dummies(data['rank'],prefix='rank')],axis=1)

```

```python
one_hot_data=one_hot_data.drop('rank',axis=1)

one_hot_data[:10]
```

```python
processed_data=one_hot_data
processed_data['gre']=processed_data['gre']/800
processed_data['gpa']=processed_data['gpa']/4
processed_data.head()
```

```python
sample=np.random.choice(processed_data.index,size=int(len(processed_data)*0.9),replace=False)
train_data,test_data=processed_data.iloc[sample],processed_data.drop(sample)

```

```python


import keras
```

```python
features=np.array(train_data.drop('admit',axis=1))
target=np.array(keras.utils.to_categorical(train_data['admit'],2))
features_test=np.array(test_data.drop('admit',axis=1))
target_test=np.array(keras.utils.to_categorical(test_data['admit'],2))

print(features[:10])
print(target[:10])
```

```python
from keras.models import Sequential
from keras.layers.core import Dense, Dropout, Activation
from keras.optimizers import SGD
from keras.utils import np_utils

model=Sequential()
model.add(Dense(128,activation='relu',input_shape=(6,)))
model.add(Dropout(.2))
model.add(Dense(64,activation='relu'))
model.add(Dropout(.1))
model.add(Dense(2,activation='softmax'))


model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
model.summary()
```

```python
model.fit(features,target,epochs=200,batch_size=100,verbose=0)
```

```python
score=model.evaluate(features,target)
```

```python
print(score[1])
```
